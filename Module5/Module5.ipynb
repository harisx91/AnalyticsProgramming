### Importing the libraries I will be working on in this project

```python
#Dependencies
import numpy as np
import pandas as pd
```

### Opening up the file saved in my project "Resources" folder

```python
#Opening up the location of the stored data
file = 'Resources/Natural_Gas_Consumption_by_ZIP_Code_-_2010.csv'
```

### Storing file in new variable using pandas library comand

```python
#Reading and using the head command to print out only first 5 rows of data
file_pd = pd.read_csv(file)
file_pd.head(10)
```

### Looking at how many rows and columns my file contain

```python
#Looking at the type of data or the data size
rows, cols = file_pd.shape
print("There are " +str(rows) + " rows and " + str(cols) + " columns in our data")
```

### Dropping unnecessary columns such as "Unnamed: 5" in my file that contains no values 

```python
#It will drop any Unnecessary columns and we will create new variable
new_file_pd = file_pd.drop(columns = ['Unnamed: 5'])
```

### Renaming columns so it's easier to read

```python
#changing the names of each column
new_file_pd = new_file_pd.rename(columns = {'Building type (service class':  "BuildingType", 'Utility/Data Source': 'Company', 
                                   ' Consumption (GJ) ': 'ConsumptionGJ'})
new_file_pd.head(10)
```

### Creating new variable, and checking to see which Building Types are in data

```python
UniqueBuildingType = new_file_pd['BuildingType'].unique()
UniqueBuildingType
```

```python
new_file_pd['BuildingType'].value_counts()
```

### In this command we will join data that is the same building type but named a little differently

```python
#Combining same building point data, and rename it to new variable
RenameLargeResidential = lambda x: x if x != 'Large residential' else 'Large Residential'
new_file_pd.BuildingType = file_pd.BuildingType.apply(RenameLargeResidential)
new_file_pd['BuildingType'].value_counts()
```

### Finding the median of all the Consumption rate in the data

```python
Median_ConsumptionGJ = new_file_pd['ConsumptionGJ'].median()
Median_ConsumptionGJ
```

### Another way to find the median

```python
new_file_pd['ConsumptionGJ'].median()
```

### Comparing the Consumption rate by each Building type

```python
compare_values = new_file_pd.groupby('BuildingType').ConsumptionGJ.median()
compare_values
```

```python
compare_values[compare_values < Median_ConsumptionGJ]
```

### Looking at the analysis of each building type, and checking to see where the values fall

```python
compare_values.groupby('BuildingType').describe()
```
