### Importing the libraries using in this project

### From the DataSet we just uploaded from DataFolder as ".arff" file; we are going to convert it to a DataFrame using pandas

```python
#Coverting the data to a dataframe
Dataone_df = pd.DataFrame(Dataone[0], dtype='i')
```

```python
#Looking at only first five rows of DataFrame
Dataone_df.head(5)
```

### Checking to see if there is some correlation I can see to futher evaluate

```python
Dataone_df.corr()
```

###  With the DataFrame we just created, we are going to save it as CSV file in "DataFolder"; referencing as Phishing and Nonphishing 

```python
phishing_df = Dataone_df[Dataone_df['Result']==1]
non_phishing_df = Dataone_df[Dataone_df['Result'] == -1]
##Save them as CSV in the DataFolder
phishing_df.to_csv('DataFolder/Phishing.csv')
non_phishing_df.to_csv('DataFolder/NonPhishing.csv')
```

### Analysis of the relationship between one or more characteristics and the outcome (phishing / non-phishing)

```python
len(phishing_df)
```

```python
len(non_phishing_df)
```

```python
phishing_mean = phishing_df.mean().sort_values(ascending = False)
```

```python
non_phishing_mean = non_phishing_df.mean().sort_values(ascending = False)
```

```python
compare_mean = pd.DataFrame([phishing_mean, non_phishing_mean]).T
compare_mean.columns = ['Phishing', 'Non Phishing']
compare_mean
```

### We can now evaluate the data of both Phishing and Non-Phishing data using mean. We can see the differences in each row. 
* SSLfinal_State: Shows highest difference between the two and suggesting that this feature is found commonly in Phishing websites. 
* Request_URL: This alse shows that this is available through Phishing websites the most


```python
compare_mean.describe()
```

### In conclussion, we can see that Phishing websites has higher mean compared to non-phishing websites. Not only that, Phishing websites has a value closer to "1" making them more likely to be a phishing website.
